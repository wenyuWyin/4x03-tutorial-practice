```julia
# create a function that takes as input a matrix size, generates a symmetric random square matrix
# with i.i.d. zero-mean Gaussian entries, and return the eigenvalues of this matrix
function part2Eigen(n)    
    A = randn(n, n);
    # triangularMatrix = [x <= y ? randn() : 0 for x in 1:n, y in 1:n];
    symmetricMatrix = Symmetric(A);
    # println(symmetricMatrix);

    eigenResult = eigen(symmetricMatrix);

    return eigenResult;
end
```

```julia
# print out the eigenvalues with highest and lowest magnitude after calling the function wuth a size of 5
size = 5;
eigenResult = part2Eigen(size);
maxEigen = maximum(abs.(eigenResult.values));
minEigen = minimum(abs.(eigenResult.values));
println("Highest magnitude: " * string(maxEigen) *" \nLowest magnitude: " * string(minEigen));
```

```julia
# call the function above N = 100 times with a size of 10
# and plot a histogram of the distribution of the eigenvalues
# (all the eigenvalues: 10*100 = 1000 numbers)
N = 100;
size = 10;

eigenValues = []
for i = 1:N
    eigenResult = part2Eigen(size);
    append!(eigenValues, eigenResult.values);
end

histogram(eigenValues, label=false)
```

```julia
# mean and standard deviation
mean = sum(eigenValues) / length(eigenValues);

sum_squared_diff = sum((x - mean)^2 for x in eigenValues);
variance = sum_squared_diff / (length(eigenValues) - 1);
std = sqrt(variance);

println("The mean of the eigen values is " * string(mean));
println("The standard deviation of the eigen values is " * string(std));
```
