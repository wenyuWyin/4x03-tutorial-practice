1. Compute the Lagrangian interpolating function for the points below:

| x  | y  |
|----|----|
| -1 | -1 |
| 0 | 2 |
| 1  | 1 |
| 1.3 | 0.967 |
| 2  | 3 |

```julia
using Plots;

x0 = -1;
x1 = 0;
x2 = 1;
x3 = 1.3;
x4 = 2;

y0 = -1;
y1 = 2;
y2 = 1;
y3 = 0.967;
y4 = 3;

L0(x) = (x-x1)*(x-x2)*(x-x3)*(x-x4)/((x0-x1)*(x0-x2)*(x0-x3)*(x0-x4));
L1(x) = (x-x0)*(x-x2)*(x-x3)*(x-x4)/((x1-x0)*(x1-x2)*(x1-x3)*(x1-x4));
L2(x) = (x-x0)*(x-x1)*(x-x3)*(x-x4)/((x2-x0)*(x2-x1)*(x2-x3)*(x2-x4));
L3(x) = (x-x0)*(x-x1)*(x-x2)*(x-x4)/((x3-x0)*(x3-x1)*(x3-x2)*(x3-x4));
L4(x) = (x-x0)*(x-x1)*(x-x2)*(x-x3)/((x4-x0)*(x4-x1)*(x4-x2)*(x4-x3));

P(x) = y0*L0(x) + y1*L1(x) + y2*L2(x) + y3*L3(x) + y4*L4(x);
```

2. Plot the given points and the interpolation function obtained above for x between -1 and 2

```julia
x_values = -1:0.001:2;
plot(x_values,P.(x_values),label="Interpolating Polynomial",xlabel="x",ylabel="y",title="Lagrange Interpolation")
scatter!([x0,x1,x2,x3,x4],[y0,y1,y2,y3,y4],label="Data Points")
```

3. Compute the maximum absolute error between the interpolated function and the function f in the range [-1, 2]. Where does it occur?
$$ f(x) = \frac{7}{6}x^3 - 2x^2 - \frac{1}{6}x + 2$$

```julia
f(x) = 7/6*x^3 -2x^2 -1/6*x + 2;
errors = abs.(f.(x_values) .- P.(x_values));
max_error = maximum(errors);
max_index = argmax(errors);
max_location = x_values[max_index];
println("The maximum error is $max_error at x = $max_location")
```
